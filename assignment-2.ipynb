```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the random variable X with 5 elements and their probabilities
X = np.array([0, 1, 2, 3, 4])
probabilities = np.array([1/1000, 2/1000, 4/1000, 2/1000, 1/1000])
probabilities = probabilities / np.sum(probabilities)  # Normalize probabilities

def sample_mean_experiment(n, num_experiments):
    samples = np.random.choice(X, size=(num_experiments, n), p=probabilities)
    sample_means = np.mean(samples, axis=1)
    return sample_means

def plot_sample_means(sample_means_list, n_values):
    for sample_means, n in zip(sample_means_list, n_values):
        plt.figure()
        sns.kdeplot(sample_means, bw_adjust=0.5)
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')
        plt.title(f'Distribution of Sample Means (n = {n})')
        plt.show()

n_values = [10, 100, 1000, 10000, 100000]  # Different values of n
num_experiments = 10000  # Number of times the experiment is repeated

sample_means_list = [sample_mean_experiment(n, num_experiments) for n in n_values]
plot_sample_means(sample_means_list, n_values)
```
